## Print the confusion matrix

For the iris dataset provided with the sklearn toy datasets library, create a logistic regression model.

After fitting the iris training dataset on this model, print the confusion matrix on the predicted test values.

Note: Use randomstate = 42 when using traintest_split function. Output:
Print the confusion matrix on the predicted test values

```python
import pandas as pd
import numpy as np
from sklearn import datasets
from sklearn.linear_model import LogisticRegression
from sklearn.metrics import confusion_matrix
from sklearn.model_selection import train_test_split

iris = datasets.load_iris()

x = iris.data
y = iris.target

x_train, x_test, y_train, y_test = train_test_split(x, y, random_state = 42)

model = LogisticRegression()
model.fit(x_train, y_train)

y_predict = model.predict(x_test)

print(confusion_matrix(y_test, y_predict))
```

## Print the Classification Report

For the iris dataset provided with the sklearn toy datasets library, create a logistic regression model.

After fitting the iris training dataset on this model, print the classification report on the predicted test values.

Note: Use randomstate = 42 when using traintest_split function. Output:
Print the classification report on the predicted test values

```python
import pandas as pd
import numpy as np
from sklearn import datasets
from sklearn.linear_model import LogisticRegression
from sklearn.metrics import classification_report
from sklearn.model_selection import train_test_split

iris = datasets.load_iris()

x = iris.data
y = iris.target

x_train, x_test, y_train, y_test = train_test_split(x, y, random_state = 42)

model = LogisticRegression()
model.fit(x_train, y_train)

y_predict = model.predict(x_test)

print(classification_report(y_test, y_predict))
```
