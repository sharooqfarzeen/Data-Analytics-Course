# Horsepower/Remove Outliers

After removing the outliers again plot the boxplot for the “highway-mpg” column and visualize if it has gotten cleaner after the outlier treatment.

Output

```python
import pandas as pd
import numpy as np
from scipy import stats

df = pd.read_csv("cars.csv")

outliers = np.abs(stats.zscore(df['highway-mpg'])) > 3
indices = df[outliers].index

df.drop(indices, inplace = True)

print(df.shape)
```

# Bidirectional Feature Selection

```python
import pandas as pd
import numpy as np
from sklearn.feature_selection import RFE, f_regression
from sklearn.linear_model import LinearRegression

df = pd.read_csv("cars.csv")

#selecting numeric column names, except response variable
cols = df.select_dtypes(include = 'number').columns[:-1]

#backward selection
lre = LinearRegression()
rfe = RFE(estimator = lre, n_features_to_select = 4, step = 1)
rfe.fit(df[cols], df['MSRP'])

cols[~rfe.support_]

#forward selection
fr = f_regression(df[cols], df['MSRP'])
p = np.array(fr[1])

cols[p < 0.05]
```

# Risk Factor

After analyzing the cars dataset, it is found out that the symbolling feature can be better studied if a better representation of risk factors is provided in the dataset. You have been assigned a task to create a new feature called "Risk Factor," which converts the numerical values of the symboling feature to categorical types.

The categories associated with the symbolling values are:

3: 'Extremely High',
2: 'Very High',
1: 'High',
0: 'Moderate',
-1: 'Safe',
-2: 'Moderalty Safe',
-3: 'Higly Safe'
Expected Output
Print the value count of the new feature "Risk Factor"

```python
import pandas as pd


df = pd.read_csv("cars.csv")

# Mapping of numerical values to categorical types
risk_factor_mapping = {3: 'Extremely High',
                       2: 'Very High',
                       1: 'High',
                       0: 'Moderate',
                       -1: 'Safe',
                       -2: 'Moderalty Safe',
                       -3: 'Higly Safe'}

# Create the new feature 'Risk Factor' based on the mapping
df['Risk Factor'] = df['symboling'].map(risk_factor_mapping)


print(df['Risk Factor'].value_counts())
```

```python
import pandas as pd

df = pd.read_csv("cars.csv")

list = ['Extremely High', 'Very High', 'High', 'Moderate', 'Safe', 'Moderalty Safe', 'Higly Safe']

df["Risk Factor"] = pd.cut(df['symboling'], bins = 7, labels = list[::-1])

df['Risk Factor'] = df["Risk Factor"].astype('category')

print(df['Risk Factor'].value_counts())
```
