As you might already know that.. 
In probability theory and statistics, variance is the expectation of the squared deviation of a random variable from its mean, and it informally measures how far a set of (random) numbers are spread out from their mean.

```python
#Let us standardize dfrtpr 
# numerical features
ncols = ['Item_Weight','Item_Visibility','Item_MRP']

# apply standardization on numerical features
for i in ncols:
    
    # fit on data columns
    scale = StandardScaler().fit(df_check[[i]])
    
    # transform the data
    df_check[i] = scale.transform(df_check[[i]])
    
# Check the variance now
#dfrtprn=dfrtpr.filter(["Item_Weight","Item_Visibility","Item_MRP"])
df_check.var()
```

It is now clear that variance was low of Item_Visibility due to multiscaling problem. Hence we cannot eliminate the Item_Visibility Dimension.
