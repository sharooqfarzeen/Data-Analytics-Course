```python
import pandas as pd

# Sample DataFrame
data = {
    'Category': ['A', 'B', 'A', 'B', 'A', 'B'],
    'Value1': [10, 20, 30, 40, 50, 60],
    'Value2': [100, 200, 300, 400, 500, 600]
}
df = pd.DataFrame(data)

# Grouping by 'Category' column and applying aggregate functions
result = df.groupby('Category').agg({
    'Value1': 'sum',      # Calculate sum of 'Value1' for each group
    'Value2': ['min', 'max']  # Calculate min and max of 'Value2' for each group
})

print(result)
```

```python
df = pd.read_csv('bigmart.csv')

df2 = df.copy()

cols = ['Item_Weight', 'Item_MRP']

from sklearn.preprocessing import StandardScaler

scaler_object = StandardScaler().fit(df2[cols])
df2[cols] = scaler_object.transform(df2[cols])

df2[cols].describe()
```
