##All Merge

Problem statement
You have been provided with two datasets:

data1 = {
  "name": ["Sally", "Mary", "John"],
  "age": [50, 40, 30]}

data2 = {
  "name": ["Sally", "Peter", "Micky"],
  "age": [77, 44, 22]
}

Your task is to merge these two datasets in such a way that all the rows of both the dataset are present in the merged dataset.

Expected output

Print the number of unique values present in the “name” column.

```python
import pandas as pd

data1 = {
  "name": ["Sally", "Mary", "John"],
  "age": [50, 40, 30]}

data2 = {
  "name": ["Sally", "Peter", "Micky"],
  "age": [77, 44, 22]}

left = pd.DataFrame(data1)
right = pd.DataFrame(data2)

union = pd.merge(left, right, how = 'outer')

print(union.name.unique())
```

##Merge Data

Problem statement
You have been given two data frames.

df1 = pd.DataFrame({'employee': ['Bob', 'Jake', 'Lisa', 'Sue'],
                'group': ['Accounting', 'Engineering', 'Engineering', 'HR']})

df2 = pd.DataFrame({'name': ['Bob', 'Jake', 'Lisa', 'Sue'],
                'salary': [70000, 80000, 120000, 90000]})
Problem Statement
You need to merge these two datasets. Here the employee for df1 is similar to the name for df2. Also, you need to remove the name column from the merged data frame.

```python
import pandas as pd

df1 = pd.DataFrame({'employee': ['Bob', 'Jake', 'Lisa', 'Sue'],
                    'group': ['Accounting', 'Engineering', 'Engineering', 'HR']})

df2 = pd.DataFrame({'name': ['Bob', 'Jake', 'Lisa', 'Sue'],
                    'salary': [70000, 80000, 120000, 90000]})

df = pd.merge(df1, df2, left_on = 'employee', right_on = 'name', how = 'inner')

df.drop('name', axis = 1, inplace = True)

print(df.shape)
```

##Filter Airline

Problem statement
The B6 airline has been struggling with handling the amount of traffic that comes to the Delhi and Banglore airports. They need to make a decision if they need to add one more aeroplane to these airports. You have been provided with the airline dataset. Your task is to filter out this dataset based on the following conditions:

1. Either the destination is DEL or the origin is BLR.
2. Air time should be less than 5 hrs.
3. Carrier should be B6.
The dataset is given below:

airline = {
    "Carrier": ["B6", "WW", "B6", "B6", "WN", "WW", "B6"],
    "Origin": ["BLR", "JAI", "SAG", "SXR", "BLR", "JLR", "DLI"],
    "Destination": ["DEL", "DEL", "HYD", "DEL", "JAI", "KLH", "RPR"],
    "Time": [3, 2, 5, 3, 6, 4, 3]
}

Expected Output

Print the sum of air_time from the filtered dataset.

```python
import pandas as pd

airline = {
    "Carrier": ["B6", "WW", "B6", "B6", "WN", "WW", "B6"],
    "Origin": ["BLR", "JAI", "SAG", "SXR", "BLR", "JLR", "DLI"],
    "Destination": ["DEL", "DEL", "HYD", "DEL", "JAI", "KLH", "RPR"],
    "Time": [3, 2, 5, 3, 6, 4, 3]
}

df = pd.DataFrame(airline)

"""Either the destination is DEL or the origin is BLR.
Air time should be less than 5 hrs.
Carrier should be B6."""

b6 = df[((df.Origin == 'BLR') | (df.Destination == 'DEL')) & (df.Time < 5) & (df.Carrier == 'B6')].copy()

print(b6.Time.sum())
```

##Highest Sales

Problem statement
For the big mart dataset, your manager wants to make a decision on opening a new outlet, she gave you the task to provide her with the data grouped by the location of the previously opened outlets and the type of outlet and find out the average itemoutletsales for each of these groups.

Expected output

Print the highest sum of sales based on the "Outlet Type" and the "Outlet Location Type".

```python
df = pd.read_csv('bigmart.csv')

df.head()
```

```python
import pandas as pd
import numpy as np
import math

df = pd.read_csv("bigmart.csv")

print(math.floor(df.groupby(['Outlet_Type', 'Outlet_Location_Type']).agg({'Item_Outlet_Sales': np.sum}).Item_Outlet_Sales.max()))
```

##Sales Data

Problem statement
An investor is interested in funding your company but first requests some sales data. Your manager has assigned you to display the records "Small" outlet size. Another requirement is that only "Tier 1" outlet locations are needed.

Expected Output

Print the highest sum of Item_Outlet_Sales grouped by "Outlet_Identifier", filtered by Outlet_Location_Type and Outlet_Size.

##Normalize Dataset

Problem statement
You have been provided with the dataset which contains some numerical columns having a different set of ranges.

But there is a problem that some of the numerical columns are not in the same range which can affect the model negatively. Now your task is to normalize the numerical features using the sklearn library.

Expected Output

Print the mean of the salary column before and after normalization in separate lines rounded off to two decimal places.

```python
import pandas as pd
from sklearn.preprocessing import MinMaxScaler

#original dataframe
df_og = pd.read_csv("customer.csv")

#printing salary mean
print(round(df_og.Salary.mean(), 2))

#new dataframe with columns that have to be normalized
df_filtered = df_og.filter(['Age', 'Salary'])

#initiating MinMaxScaler object, and finding minimum and maximum of required columns using fit()
min_max_object = MinMaxScaler().fit(df_filtered)

#normalized array
norm_array = min_max_object.transform(df_filtered)

#replacing non-normalized columns in OG dataframe with normalized columns
df_og[df_filtered.columns] = norm_array

#printing normalized salary mean
print(round(df_og.Salary.mean(), 2))
```
