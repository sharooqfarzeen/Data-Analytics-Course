## Data Normalization

**Normalization is a scaling technique in which values are shifted and rescaled so that they end up ranging between 0 and 1. It is also known as Min-Max scaling.**


```python
# data normalization with sklearn
from sklearn.preprocessing import MinMaxScaler
```

```python
# let us seperate out numerical features from dataset

dfrt2 = df.copy()
dfnorm=dfrt2.filter(['Item_Weight','Item_MRP'])
```

```python
# fit scaler on data
norm = MinMaxScaler().fit(dfnorm)
print(norm)
```

```python
# transform the data
dfnorm1 = norm.transform(dfnorm)
```

```python
# extract rowno as ndarray from dfrt2 to add to dfnorm1 ndarray
rownoarr=dfrt2[['rowno']].to_numpy()
```

```python
# append dfnorm2 ndarray with rowno and convert the resulting array into dataframe
dfnorm2=pd.DataFrame(np.append(dfnorm1,rownoarr,axis=1), columns=['Item_Weight','Item_MRP','rowno'])
```

```python
# drop non-normalized columns from original data frame dfrt2 and store in new data frame dfr2.
dfr2=dfrt2.drop(['Item_Weight','Item_MRP'],axis=1)
```

```python
#megre the two data frames dfnorm2 and dfr2 joining on 'rowno')
dfrt2=pd.merge(dfr2, dfnorm2,on='rowno')
```
