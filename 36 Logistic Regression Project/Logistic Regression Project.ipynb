## Importing required modules

```python
import pandas as pd
import numpy as np
from sklearn.model_selection import train_test_split
from sklearn import datasets
from sklearn.linear_model import LogisticRegression
```

## Loading Dataset

```python
train = pd.read_csv("0000000000002429_training_titanic_x_y_train.csv")
test = pd.read_csv("0000000000002429_test_titanic_x_test.csv")
```

## Analyzing Data

```python
train.head()
```

```python
train.describe()
```

```python
train.info()
```

```python
test.head()
```

## Handling Null values

```python
train['Age'].fillna(train['Age'].median(), inplace = True)
test['Age'].fillna(train['Age'].median(), inplace = True)
```

## Encoding Categorical variables to numeric variables

```python
gender_maps = {'female': 1,
               'male': 2}

embarked_maps = {'C': 1,
                 'Q': 2,
                 'S': 3}

train['Sex'] = train['Sex'].map(gender_maps)
train['Embarked'] = train['Embarked'].map(embarked_maps)

test['Sex'] = test['Sex'].map(gender_maps)
test['Embarked'] = test['Embarked'].map(embarked_maps)
```

```python
train['Embarked'].mode()
```

```python
train['Embarked'].fillna(3.0, inplace = True)
test['Embarked'].fillna(3.0, inplace = True)
```

```python
train.info()
```

```python
train.head()
```

## Splitting Training data to features and target variable

```python
x_train = train.iloc[:, [0, 2, 3, 4, 5, 7, 9]]
y_train = train.iloc[:, -1]

x_test = test.iloc[:, [0, 2, 3, 4, 5, 7, 9]]
```

```python
x_train.head()
```

```python
y_train.head()
```

```python
x_train['Sex_Sex'] = x_train['Sex']**2
x_test['Sex_Sex'] = x_train['Sex']**2
```

## Initializing and Training Model

```python
model = LogisticRegression(C = 2, multi_class = 'multinomial')
model.fit(x_train, y_train)
```

## Predicting for test data

```python
prediction = model.predict(x_test)
```

```python
prediction
```

```python
model.score(x_train, y_train)
```

## Saving predictions to CSV file

```python
np.savetxt("prediction1.csv", prediction, fmt='%.0f', delimiter = ",")
```
