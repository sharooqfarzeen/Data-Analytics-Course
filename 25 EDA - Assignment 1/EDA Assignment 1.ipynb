## Missing Indicator

```python
import pandas as pd
import numpy as np

df = pd.read_csv("cars.csv")

df.fillna(np.NaN, inplace = True)

df.replace('?', np.NaN, inplace = True)

print(df.isnull().sum())
```

## Maker

```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt

df = pd.read_csv('cars.csv')

df.make.value_counts().nlargest(10).plot(kind='bar', figsize=(15,5))
plt.title("Number of vehicles by manufacturer")
plt.ylabel('Number of vehicles')
plt.xlabel('Manufacturer')
plt.show()

print(df.make.value_counts().index[0])
```

## Ratings

```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt

df = pd.read_csv("cars.csv")

df.symboling.value_counts().plot(kind = 'bar', figsize = (15,5))
plt.title("Insurance Risk Category")
plt.ylabel("Number of cars")
plt.xlabel("Insurance Risk Category")
plt.show()

print(df.symboling.value_counts().index[0])
```

## Aspiration Type

```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt

df = pd.read_csv("cars.csv")

df.aspiration.value_counts().plot(kind = 'pie', figsize = (15,5))
plt.show()

print(df.aspiration.value_counts().index[0])
```
